```python
#importing all the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('MatNavi Mechanical properties of low-alloy steels.csv')
```

```python
#tensile strenght as target variable

plt.figure(figsize=(8,6))
sns.histplot(Data[' Tensile Strength (MPa)'], bins = 25)
plt.title('Tensile Stength Distribution')
plt.xlabel('Material Tensile Strength (MPa)')
plt.ylabel('Number of readings')
plt.show()
```

```python
print(Data.columns)
```

```python
Data.columns = Data.columns.str.strip()
print(Data.columns)
```

```python
#variation with composition

composition_of_elements = ['C',	'Si',	'Mn',	'P',	'S',	'Ni',	'Cr',	'Mo',	'Cu',	'V',	'Al',	'N',	'Ceq',	'Nb + Ta','Temperature (°C)']
correlation = Data[composition_of_elements+['Tensile Strength (MPa)']].corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, annot=True, center=0, cmap='coolwarm')
plt.title('Tensile Strength correlation with Elements')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))
sns.scatterplot(x='C', y='Tensile Strength (MPa)', data=Data, ax=axes[0])
axes[0].set_title('Carbon % vs Tensile Strength')

sns.scatterplot(x='Ceq', y='Tensile Strength (MPa)', data=Data, ax=axes[1])
axes[1].set_title('Carbon Equivalent vs Tensile Strength')
plt.show()
```
