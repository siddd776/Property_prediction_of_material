```python
#histogram to see properties relation with element composition

target_columns = ['0.2% Proof Stress (MPa)', 'Tensile Strength (MPa)', 'Elongation (%)',
       'Reduction in Area (%)']
fig, axes = plt.subplots(2,2, figsize = (12,10))
axes = axes.flatten()

for i, col in enumerate(target_columns):
    sns.histplot(Data[col], kde = True, bins=30, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')

plt.show()
```

```python
#scatter plot of composition of elemets with tensile strength

composition_cols = ['C', 'Si', 'Mn', 'P', 'S', 'Ni', 'Cr', 'Mo', 'Cu', 'V',
       'Al', 'N', 'Ceq', 'Nb + Ta']
fig,axes = plt.subplots(4,4, figsize=(18,20))
axes = axes.flatten()

for i,col in enumerate(composition_cols):
    sns.scatterplot(x=Data[col], y=Data['Tensile Strength (MPa)'], ax=axes[i])
    axes[i].set_title(f'{col} vs Tensile Strength')

plt.show()
```

```python
#data split using group shuffle split
from sklearn.model_selection import GroupShuffleSplit

group_shuffle_shift =  GroupShuffleSplit(n_splits= 1, test_size=0.25, random_state=42)
train_index, test_index = next(group_shuffle_shift.split(Data, groups=Data['Alloy code']))

train_Data = Data.iloc[train_index]
test_Data = Data.iloc[test_index]
```

```python
#checking multicollinearity using VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

x = Data[composition_cols].copy()

vif_data =  pd.DataFrame()
vif_data['Feature'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif_data =  vif_data.sort_values('VIF', ascending=False)
print(vif_data)
```

```python
#data shows high multicollinearity
#using ridge regression

feature_col = ['C', 'Si', 'Mn', 'P', 'S', 'Ni', 'Cr', 'Mo', 'Cu', 'V',
       'Al', 'N', 'Nb + Ta']
#excluding ceq because it is derived using formula

x_train =  train_Data[feature_col]
y_train =  train_Data['Tensile Strength (MPa)']

x_test = test_Data[feature_col]
y_test = test_Data['Tensile Strength (MPa)']
```

```python
#scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled =  scaler.fit_transform(x_train)
x_test_scaled =  scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

ridge= Ridge(alpha=1.0, random_state=42)
ridge.fit(x_train_scaled, y_train)
ridge_predict = ridge.predict(x_test_scaled)

print("MAE is ", mean_absolute_error(y_test, ridge_predict))
print("RMSE is ", np.sqrt(mean_squared_error(y_test, ridge_predict)))
print("r2 score is ", r2_score(y_test, ridge_predict))
```

```python
#random forest regresion

from sklearn.ensemble import RandomForestRegressor

random_forest =  RandomForestRegressor(n_estimators=200, random_state=42)
random_forest.fit(x_train,y_train)
random_forest_predict = random_forest.predict(x_test)

print("MAE is ", mean_absolute_error(y_test, random_forest_predict))
print("RMSE is ", np.sqrt(mean_squared_error(y_test, random_forest_predict)))
print("r2 score is ", r2_score(y_test, random_forest_predict))
```
